####**3) Identificar cada característica como numérica, ordinal o nominal**

**Ordinales:**

1) Una característica ordinal que se podría presumir es si la persona es fumadora o no, siendo 0 cuando no lo es y 1 cuando sí lo es, dado que sí hay mayor riesgo asociado con esta condición, además, los cargos por tratamiento podrían ser mayores si se requiere algún tipo de cuidado especial.

2) Dependiendo de la estuctura del plan  de salud, la cantidad de hijos podría considerarse como una características ordinal, dado que podrían darse variación en los cargos de determinadas situaciones que sí sean dependientes del número de hijos, por lo que se daría una diferenciación.

**Nominales:**

1) La region donde se encuentra considero que es nominal incialmente, dado que no conozco el tipo de plan de salud por lo que no podría decir cuál ciudad tomaría mayor relevancia, igualmente, tener tarifas tan variables para pretadores únicos podría acarrear dificultades contables.

2) Desde mi punto de vista no debería haber prelación entre hombres y mujeres en los cargos, pero los podría haber, por mi consideración propia, lo voy a poner como un nominal.

**Numéricas:**

1) El índice de masa corporal.

2) Los cargos

3) Edad. La edad individual no creo que sea ordinal, grupos o rangos de edad sí lo son; por tanto, la edad no la tomaría ordinal, si hubiera un grupo asociado a rangos de edades posiblemente sí lo tomaría ordinal, en esta caso no por que diríamos que cada año que pasa en la vida de una persona sus cargos podrían varias y creo que un sistema no soporta tal variación tarifaria.


####**4) Codificar de manera ordinal cualquier característica ordinal**

```python
#Dado que la característica de fumador o no se considera ordinal y va a estar en el conjunto de características, se va a cambiar antes de realizar la división de conjunto
# de entrenamiento y de pruebas
data_insurance['smoker'] = data_insurance['smoker'].replace({'yes':1,'no':0})
```

####**5) Definir las características (X) y el objetivo (y)**

```python
#El vector objetivo es charges
y = data_insurance['charges']
X=data_insurance.drop(columns='charges')
```

```python
y
```

```python
X
```

####**6) Realizar un train test split a los datos para prepararlos para el aprendizaje automático**

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
X_train, X_test, y_train, y_test
```

####**7) Realizar una codificación one-hot cualquier característica nominal**

```python
data_selector_object = make_column_selector(dtype_include='object')
```

```python
#Selección de columnas categóricas de entranmiento
Xtrain_category_data = X_train[data_selector_object(X_train)]
Xtrain_category_data
```

```python
#Selección de las columnas categóricas del conjunto de prueba
Xtest_category_data = X_test[data_selector_object(X_test)]
Xtest_category_data
```

```python
#OneHotEncoder
#1- Instanciamos la codificación OneHotEnconder
ohe_enconder = OneHotEncoder(sparse=False, handle_unknown='ignore')

#2- Ajustar el OneHotEncoder en los datos de entrenamiento pero solo para el dataframe de variables categóricas, evite la fuga de datos
ohe_enconder.fit(Xtrain_category_data)

#3- Transformar los datos de entrenamiento y los de prueba
Xtrain_ohe = ohe_enconder.transform(Xtrain_category_data)
Xtest_ohe = ohe_enconder.transform(Xtest_category_data)
```

```python
Xtrain_ohe
```

```python
#Llevar a dataframe las variables categóricas que ya tenemos
name_column_category = ohe_enconder.get_feature_names_out(Xtrain_category_data.columns)
name_column_category
```

```python
# Conformación de los df
Xtrain_ohe = pd.DataFrame(Xtrain_ohe,columns=name_column_category)
Xtest_ohe = pd.DataFrame(Xtest_ohe, columns=name_column_category)
```

```python
Xtrain_ohe
```

```python
Xtest_ohe
```

####**8) Escalar cualquier característica numérica**

```python
data_selector_number = make_column_selector(dtype_include='number')
#Selección de columnas categóricas de entranmiento
Xtrain_number_data = X_train[data_selector_number(X_train)]
Xtrain_number_data
```

```python
#Selección de columnas categóricas de entranmiento
Xtest_number_data = X_test[data_selector_number(X_test)]
Xtest_number_data
```

```python
#Creador del Scaler
scaler = StandardScaler()
# Ajuste del escaler de los datos de entrenamiento
scaler.fit(Xtrain_number_data)
```

```python
#Tranformación de los datos de entrenamiento
Xtrain_scaled = scaler.transform(Xtrain_number_data)
Xtrain_scaled
```

```python
#Tranformación de los datos de prueba
Xtest_scaled = scaler.transform(Xtest_number_data)
Xtest_scaled
```

```python
#Creamos nuevos dataframe para los datos de entrenamiento y los de prueba
name_columns_number = scaler.get_feature_names_out(Xtrain_number_data.columns)
name_columns_number
```

```python
#DatFrame de los datos de entrenamiento
Xtrain_scaled = pd.DataFrame(Xtrain_scaled,columns=name_columns_number)
Xtrain_scaled
```

```python
# DataFrame de los datos de prueba
Xtest_scaled = pd.DataFrame(Xtest_scaled,columns=name_columns_number)
Xtest_scaled
```

####**9) Concatenar todas las características de vuelta a un DataFrame**

```python
#DataFrame de datos de entranamiento escalados y aplicado OneHotEncoder
X_train_concat = pd.concat([Xtrain_ohe,Xtrain_scaled], axis=1)
X_train_concat
```

```python
#DataFrame de datos de prueba escalados y aplicado OneHotEncoder
X_test_concat = pd.concat([Xtest_ohe,Xtest_scaled], axis=1)
X_test_concat
```
