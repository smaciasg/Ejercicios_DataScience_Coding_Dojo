####**3) Crear matriz de características y vector objetivo**

```python
X = data_abalones.drop(columns = 'Rings')
y = data_abalones['Rings']
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

####**4) Crear los selectores de columnas categóricas y numéricas**

```python
number_selector = make_column_selector(dtype_include='number')
category_selector = make_column_selector(dtype_include='object')
```

####**5) Instanciar los transformadores numérico y categórico**

```python
# Transformador numérico
scaler = StandardScaler()
# Transformador categórico
ohe = OneHotEncoder(handle_unknown='ignore')
```

####**6) Crear tuplas entre el transformador y los selectores**

```python
number_tuple = (scaler, number_selector)
category_tuple = (ohe, category_selector)
```

####**7) Instanciar el ColumnTransformer**

```python
col_transformer = make_column_transformer(number_tuple,category_tuple, remainder='passthrough')
```

####**8) Ajustar el transformador - SOLO SE AJUSTAN LOS DATOS DE ENTRENAMIENTO**

```python
col_transformer.fit(X_train)
```

####**9) Transformación de los datos de entrenamiento y de los de prueba**

```python
X_train_processed = col_transformer.transform(X_train)
X_test_processed = col_transformer.transform(X_test)
```

####**10) Crear DataFrames para ver la transformación de los datos de entrenamiento y de los de prueba**

```python
X_train_df = pd.DataFrame(X_train_processed)
X_train_df
```

```python
X_test_df = pd.DataFrame(X_test_processed)
X_test_df
```
