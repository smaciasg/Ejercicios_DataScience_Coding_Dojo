#####**GRAFICADO MÚLTIPLE**

```python
df_undp_indices['Gender'].unique()
```

```python
lista_indicadores = ['Human Development Index', 'Life Expectancy at Birth',
       'Expected Years of Schooling', 'Mean Years of Schooling',
       'Gross National Income Per Capita']

lista_anios = df_undp_indices['Year'].unique()

for i in lista_indicadores:
  ####### PROMEDIOS MEAN #######
  #Vamos a crear un gropby para cada uno de los índices, agrupando por país y por género
  df_year_total= df_undp_indices[df_undp_indices['Gender']=='Total'].groupby(['Year'])[i].mean()
  eje_anios = df_year_total.index
  #Vamos a crar un groupby solo para las mujeres
  df_year_females= df_undp_indices[df_undp_indices['Gender']=='Female'].groupby(['Year'])[i].mean()
  #Vamos a crar un groupby solo para las hombres
  df_year_male= df_undp_indices[df_undp_indices['Gender']=='Male'].groupby(['Year'])[i].mean()

  ####### MEDIANA ###########
  #Calculamos la mediana para el total y poder comparar lo que realmente ocurre en la mayoría de los casos
  df_year_total_median= df_undp_indices[df_undp_indices['Gender']=='Total'].groupby(['Year'])[i].median()

  ###### GRÁFICA ######
  fig,axes = plt.subplots(1,1, figsize=(13,5))
  axes.plot(eje_anios, df_year_total.values,marker='o', markersize=5,c='#FEBD04', label=f'mean-{i}-total')
  axes.plot(eje_anios, df_year_total_median.values,marker='s', markersize=5,c='#015E5F', label=f'median-{i}-total')
  axes.plot(eje_anios,df_year_females.values,marker = 'o',markersize=2, c='#EF4C53', label=f'mean-{i}-female')
  axes.plot(eje_anios,df_year_male.values,marker = 'o',markersize=2, c='#566ACF', label=f'mean-{i}-male')
  axes.set_xlabel('Años')
  axes.set_ylabel('Valor medio indicador')
  axes.set_title(f'Valor promedio de {i} entre los años {lista_anios.min()}-{lista_anios.max()}')
  axes.legend()
  plt.xticks(eje_anios,rotation=90) #Poner la variable me permite poner todas las etiquetas
  print('---------------------------------------------------------------------------------------------------------------------------------------')
  plt.show()
```

**POR PAÍS**

```python
df_undp_indices['Country'].unique()
```

```python
lista_indicadores = ['Human Development Index', 'Life Expectancy at Birth',
       'Expected Years of Schooling', 'Mean Years of Schooling',
       'Gross National Income Per Capita']

for i in lista_indicadores:
  ####### PROMEDIOS MEAN #######
  #Vamos a crear un gropby para cada uno de los índices, agrupando por país y por género
  df_country_total_mean= df_undp_indices[df_undp_indices['Gender']=='Total'].groupby(['Country'])[i].mean().sort_values(ascending=False).head(50)

  ####### MEDIANA ###########
  #Calculamos la mediana para el total y poder comparar lo que realmente ocurre en la mayoría de los casos
  df_country_total_median= df_undp_indices[df_undp_indices['Gender']=='Total'].groupby(['Country'])[i].median().sort_values(ascending=False).head(50)

  ###### GRÁFICA ######
  fig,axes = plt.subplots(1,2, figsize=(20,5))
  axes[0].scatter(df_country_total_mean.index, df_country_total_mean.values,marker='o',c='#FEBD04', label=f'mean-{i}-total | entre los años {lista_anios.min()}-{lista_anios.max()}')
  axes[1].scatter(df_country_total_median.index, df_country_total_median.values,marker='s',c='#015E5F', label=f'median-{i}-total | entre los años {lista_anios.min()}-{lista_anios.max()}')
  axes[0].set_xlabel('País')
  axes[1].set_xlabel('País')
  axes[0].set_ylabel('Valor medio indicador')
  axes[1].set_ylabel('Valor medio indicador')
  axes[0].set_title(f'Valor promedio de {i} para cada país')
  axes[1].set_title(f'Valor mediana de {i} para cada país')
  axes[0].legend()
  axes[1].legend()
  axes[0].set_xticks(df_country_total_mean.index) #Poner la variable me permite poner todas las etiquetas
  axes[0].set_xticklabels(df_country_total_mean.index, rotation=90, ha='right')
  axes[1].set_xticks(df_country_total_median.index) #Poner la variable me permite poner todas las etiquetas
  axes[1].set_xticklabels(df_country_total_median.index, rotation=90, ha='right')

  # plt.xticks(rotation=90)
  print('---------------------------------------------------------------------------------------------------------------------------------------')
  plt.show()
```
